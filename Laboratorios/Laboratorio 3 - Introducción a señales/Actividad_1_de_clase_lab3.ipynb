```python
import numpy as np
from scipy import signal #fisicos, electrónica, estadística --> signal es para el filtrado y trabajo de señales (frq vs time)
import matplotlib.pyplot as plt
```

```python
A = 5
phi = 0
f = 100
w = 2*np.pi*f

fs = 20*f
Ts = 1/fs
n = np.arange(1000)
t = n*Ts

x = A*np.sin(w*t + phi)
```

```python
plt.plot(t,x)
plt.xlim(0,0.1)
```

```python
plt.stem(n,x)
plt.xlim(0,100)
```

```python
plt.magnitude_spectrum(x, Fs=fs, scale='dB')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud (dB)')
plt.grid(ls=":")
```

```python
f, t, Zxx = signal.stft(x, fs, nperseg=100)

# Graficar el espectrograma
plt.figure(figsize=(15, 4))
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('Espectrograma de la STFT')
plt.ylabel('Frecuencia [Hz]')
plt.xlabel('Tiempo [s]')
plt.colorbar(label='Magnitud')
plt.show()
```
