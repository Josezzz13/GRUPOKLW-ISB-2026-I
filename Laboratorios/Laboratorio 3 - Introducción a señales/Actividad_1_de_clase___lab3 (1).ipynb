# Generar una señal senoidal y mostrar sus gráficas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy import signal
```

## Ejercicio:

Crear 3 señales entre senoidales y cosenoidals y graficar en frecuencia

```python
#señales sinusoidales
#Definir la frecuencia de muestreo
fs = 1000*2*10 # min para aplicar Nyquist fs>=2*fmax
#--> la fmax es la mayor en la ecuacion compuesta por las señales
Ts = 1/fs
n = np.arange(1000)
t = n*Ts

# Creación de las señales
A = 1
f = 250          #  hz
w = 2*np.pi*f     #  rad/s
phi = 0

#
A1= 10
f1 = 100          #  hz
w1 = 2*np.pi*f1     #  rad/s
phi1 = np.pi/2
#
A2= 20
f2 = 1000      #  hz
w2 = 2*np.pi*f2     #  rad/s
phi2 = np.pi/4

# señal senoidales y cosenoidales
x = A*np.sin(w*t + phi)
x1= A1*np.cos(w1*t+phi1)
x2= A2*np.sin(w2*t+phi2)
xf=x+x1+x2

plt.plot(t,xf);
plt.xlim(0, 0.05)
plt.show()

plt.stem(n,xf)
plt.xlim(0, 200)
plt.show()

f, t, Zxx = signal.stft(xf, fs, nperseg=1024)

# Graficar el espectrograma
plt.figure(figsize=(15, 4))
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('Espectrograma de la STFT')
plt.ylabel('Frecuencia [Hz]')
plt.xlabel('Tiempo [s]')
plt.ylim(0, 1500)
plt.colorbar(label='Magnitud')
plt.show()

#Grafica de la frecuencia
plt.magnitude_spectrum(xf, Fs=fs, scale='dB')
plt.title('Gráfica de las frecuencias')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud (dB)')
plt.grid(ls=":")
plt.show()
```
