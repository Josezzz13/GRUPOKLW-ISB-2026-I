```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from opensignalsreader import OpenSignalsReader

# Ruta del archivo subido
archivo = '/content/E2-D1.txt'

# Leer archivo
ekg = OpenSignalsReader(archivo);

# Extraer señal del canal 2
ekg_signal = ekg.signal(2);

# Frecuencia de muestreo
fs = 1e3  # Hz

# Limitar tiempo
idx_10 = int(10*fs)
idx_38 = int(38*fs)

# Crear vector de tiempo
N = len(ekg_signal)
tiempo = np.arange(N) / fs

tiempo_basal = tiempo[idx_10:idx_38]
ekg_signal_basal = ekg_signal[idx_10:idx_38]

# remove Dc components
ekg_basal_centrado = ekg_signal_basal - np.mean(ekg_signal_basal)

# en función de la frecuencia
freq_basal = np.fft.rfftfreq(len(ekg_signal_basal), d=1/fs)
esp_basal = np.abs(np.fft.rfft(ekg_basal_centrado))

# Graficar
plt.figure(figsize=(15,5))
plt.suptitle('Señal ECG - D1')
plt.plot(tiempo_basal, ekg_signal_basal)
plt.title('Dominio del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(15,5))
plt.plot(freq_basal, esp_basal)
plt.title('Dominio de la frecuencia')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(15,5))
plt.plot(tiempo_basal, ekg_signal_basal)
plt.gca().invert_yaxis()

plt.xlim(10, 14)
plt.title('Dominio del tiempo - Zoom [10 14]')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.tight_layout()
plt.show()
```
